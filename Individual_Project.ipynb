## **Step 1**

Ask users for input. Well start writing our program at the point where the human enters the numbers that they would like the computer to work with. For this program, we would like the user to input two numbers, so lets have the program ask for two numbers. When asking for input, we should include a space at the end of our string so that there is a space between the users input and the asked string.

```python
#Ask user for two inputs
import math
def main ():
  number_1 =float(input("Enter the first number: "))
  number_2 = float(input("Enter the second number: "))
  return number_1, number_2

number_1, number_2 = main()
operation = input("Please type the operation that you would like to complete (+, -, *, /, ): ")
```

## **Step 2**

By using `try` and `except` statement, write down a function to determine whether input numbers are integers or decimals.

```python
def number_type(number):
    try:
        float_number = float(number)
        if float_number % 1 == 0:
            return int(float_number)
        else:
            return float_number
    except ValueError:
        print("Invalid input! Please enter a valid number.")
        return None

num_1 = number_type(number_1)
num_2 = number_type(number_2)
print(type(num_1), type(num_2))
```

## **Step** 3

Well name the function `calculate()` and add an additional layer of indentation within the function itself. Function `calculate()` is defining arithmetic operations.

```python
class Calculator:
    def calculate(self, num_1, num_2, operation):

        #addition
        if operation == '+':
            return num_1 + num_2

        #subtraction
        elif operation == '-':
            return num_1 - num_2

        #multiplication
        elif operation == '*':
            return num_1 * num_2

        #division
        elif operation == '/':
            if num_2 == 0:
                return "Error! Division by zero."
            else:
                return num_1 / num_2

        #square root
        elif operation == '':
            if num_1 < 0:
                return "Error! Square root of a negative number."
            else:
                return math.sqrt(num_1)


        else:
            return "Invalid Operation!"

calc = Calculator()
result = calc.calculate(num_1, num_2, operation)
print("Result:", result)
```

## Step 4

Define an `again()` function to ask user either he/she wants to do a a calculation again or not.

```python
#
def again(calc_again, previous):
    if calc_again.upper() == 'Y':
        return previous
    elif calc_again.upper() == 'N':
        print('See you later.')
        return None
    else:
        print('Invalid input. Please enter Y or N.')
        return None

again = input("Do you want to calculate again? (Y/N): ")
```

## **Step 5**



Write a `while` loop to operate the calculator until it says No and `break` the loop.

```python
while True:
    num_1, num_2 == main()
    if num_1 is None or num_2 is None:
        continue
    operation = input("Please type the operation that you would like to complete (+, -, *, /, ): ")
    calc = Calculator()
    result = calc.calculate(num_1, num_2, operation)
    print("Result:", result)

    calc_again = input("Do you want to calculate again? (Y/N): ")
    if again(calc_again, (num_1, num_2)) is None:
        break
```
